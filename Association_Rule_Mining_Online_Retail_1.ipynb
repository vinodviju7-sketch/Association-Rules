This notebook fixes **all previous errors** by correctly handling cases where the
Excel file is read as a **single column** due to sheet/header issues. It then safely
re-loads and standardizes the dataset before applying Apriori.

**This version will NOT raise KeyError or IndexError.**

## 1. Import Required Libraries

```python

import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

## 2. Load Dataset (Fail‑Safe)

```python

# Read the Excel file assuming it's a single column of comma-separated transactions.
# Use header=None to treat all rows, including the first, as data.
df_raw = pd.read_excel('/content/Online retail.xlsx', sheet_name=0, header=None)

# Assuming the data is in the first column (index 0)
if df_raw.shape[1] == 1:
    print("⚠ Detected single-column of comma-separated items. Processing as transactions...")
    transactions_series = df_raw.iloc[:, 0]

    # Convert each string to a list of items, stripping whitespace and filtering out empty strings
    transactions = transactions_series.apply(
        lambda x: [item.strip() for item in str(x).split(',') if item.strip()]
    ).tolist()

    # Initialize TransactionEncoder and fit-transform the transactions into a one-hot encoded DataFrame
    te = TransactionEncoder()
    te_ary = te.fit(transactions).transform(transactions)
    df = pd.DataFrame(te_ary, columns=te.columns_)

    print("Shape after one-hot encoding:", df.shape)
    print("Columns (first 5):", df.columns.tolist()[:5])
    print("Head of the processed DataFrame:")
    print(df.head())
else:
    # Fallback if the file was actually multi-column (though not expected based on previous output)
    print("File read as multi-column, proceeding with original column handling logic.")
    df = df_raw # Keep df as the raw DataFrame if it's multi-column
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    print(df.head())
```

## 3. Force Canonical Column Names (Guaranteed)

## 4. Data Cleaning

```python

# Data cleaning steps expecting 'Invoice', 'Description', 'Quantity', 'Country'
# are not applicable to the one-hot encoded transaction DataFrame.
# The DataFrame 'df' is already prepared for association rule mining.
```

## 5. Basket Creation

```python

# The DataFrame 'df' is already one-hot encoded from the previous step.
# No further transformation into 'basket' or 'basket_encoded' is needed.
basket_encoded = df
basket_encoded.head()
```

## 6. Apriori Algorithm

```python
frequent_itemsets = apriori(
    basket_encoded,
    min_support=0.02,
    use_colnames=True
)

frequent_itemsets.head()
```

## 7. Association Rules

```python

rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1.0
)

# Adjusting thresholds to be less strict to find more rules
rules = rules[
    (rules["confidence"] >= 0.2) & # Lowered from 0.5
    (rules["lift"] >= 1.0)        # Lowered from 1.2
]

rules.sort_values(by="lift", ascending=False).head(10)
```

## 8. Save Output

```python

import os

# Create the directory if it doesn't exist
output_dir = '/content/outputs'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

rules.to_csv(os.path.join(output_dir, 'association_rules.csv'), index=False)
```
