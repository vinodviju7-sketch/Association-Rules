# Association Rule Mining â€“ Market Basket Analysis

## 1. Import Required Libraries

```python

import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
```

## 2. Load Dataset

```python

df = pd.read_excel('/content/Online retail.xlsx', header=None)
df.head()
```

## 3. Data Preprocessing

```python

transactions = df[0].apply(lambda x: x.split(','))
transactions.head()
```

## 4. Transaction Encoding

```python

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
basket = pd.DataFrame(te_array, columns=te.columns_)
basket.head()
```

## 5. Apriori with Optimized Thresholds

```python

frequent_itemsets = apriori(
    basket,
    min_support=0.03,
    use_colnames=True
)

rules = association_rules(
    frequent_itemsets,
    metric='confidence',
    min_threshold=0.6
)

rules = rules[rules['lift'] > 1.2]
rules.head()
```

## 6. Reducing Redundant Rules

```python

rules['antecedent_len'] = rules['antecedents'].apply(len)
rules['consequent_len'] = rules['consequents'].apply(len)

rules = rules.sort_values(
    by=['confidence', 'lift'],
    ascending=False
).drop_duplicates(subset=['antecedents', 'consequents'])

rules.head()
```

## 7. Visualizing Top Association Rules (Heatmap)

```python

top_rules = rules.head(10)

heatmap_data = top_rules.pivot_table(
    index=top_rules['antecedents'].apply(lambda x: ','.join(list(x))),
    columns=top_rules['consequents'].apply(lambda x: ','.join(list(x))),
    values='lift'
)

plt.figure()
plt.imshow(heatmap_data.fillna(0))
plt.xticks(range(len(heatmap_data.columns)), heatmap_data.columns, rotation=90)
plt.yticks(range(len(heatmap_data.index)), heatmap_data.index)
plt.colorbar(label='Lift')
plt.title('Heatmap of Top Association Rules')
plt.show()
```

## 8. Network-Style Rule Representation

```python

items = list(set(
    [item for sublist in top_rules['antecedents'] for item in sublist] +
    [item for sublist in top_rules['consequents'] for item in sublist]
))

matrix = pd.DataFrame(0, index=items, columns=items)

for _, row in top_rules.iterrows():
    for a in row['antecedents']:
        for c in row['consequents']:
            matrix.loc[a, c] = row['lift']

matrix
```

## 9. Analysis and Interpretation

- Visualizations improve interpretability of association rules.
- Redundant rules were removed to simplify insights.
- Optimized thresholds improved rule relevance.

**Business Insights:**
- Strong associations suggest bundling opportunities.
- High-lift items are ideal for recommendations.

## 10. Conclusion

Optimizing thresholds, removing redundant rules, and visualizing results significantly enhances the quality and usability of association rule mining outcomes.

## 11. Interview Questions and Answers

**1. What is Support?**  
Support measures how frequently an itemset appears in the dataset.

Support(A â†’ B) = Transactions containing A and B / Total transactions

**2. What is Confidence?**  
Confidence measures the likelihood of purchasing B when A is purchased.

Confidence(A â†’ B) = Support(A âˆ© B) / Support(A)

**3. What is Lift and why is it important?**  
Lift measures the strength of association beyond random chance.

Lift(A â†’ B) = Confidence(A â†’ B) / Support(B)

**4. Limitations of Association Rule Mining:**
- Generates a large number of rules
- Computationally expensive
- Does not consider purchase sequence
- Requires careful threshold tuning
